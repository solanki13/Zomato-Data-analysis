```python
import pandas as pd
df_z=pd.read_csv('zomato.csv',encoding='ISO-8859-1')
df_z.head()
```

```python
uploaded=files.upload()
df_c=pd.read_excel('Country-Code.xlsx')
df_c.head()
```

```python
print(df_z.columns)
print(df_c.columns)
print(df_z.shape)
print(df_c.shape)
```

```python
df_merged=pd.merge(df_z,df_c,on='Country Code',how='left')
df_merged.head()
```

```python
restaurant_counts = df_merged.groupby('Country')['Restaurant Name'].count().reset_index()
restaurant_counts.columns = ['Country', 'Number of Restaurants']
restaurant_counts = restaurant_counts.sort_values(by='Number of Restaurants', ascending=False)
restaurant_counts  # Top 10 countries

```

```python
table_booking = df_merged.groupby('Has Table booking')['Restaurant Name'].count().reset_index()
table_booking.columns = ['Has Table Booking', 'Number of Restaurants']
table_booking
```

```python
TableBooking_Yes = df_merged[(df_merged['Has Table booking'] == 'Yes')& (df_merged['Country']=='India')]
restraurant_name = TableBooking_Yes['Restaurant Name']
restraurant_name
```

```python
online_delivery = df_merged.groupby("Has Online delivery")['Restaurant Name'].count()
online_delivery.columns = ['Has Online Delivery', 'No. of Restaurants']
online_delivery
```

```python
onlineDel_Yes = df_merged[(df_merged['Has Online delivery']=='Yes')&(df_merged['Country']=='India')]
restraurant_names = onlineDel_Yes['Restaurant Name']
#restraurant_names
unique_names = pd.Series(restraurant_names.unique()).reset_index(drop=True)
unique_names
```

```python
india_df = df_merged[df_merged['Country'] == 'India']

india_df = india_df.dropna(subset=['Aggregate rating', 'Votes'])

top_rated = india_df.sort_values(
    by=['City', 'Aggregate rating', 'Votes'], ascending=[True, False, False]
)
top_in_each_city = top_rated.groupby('City').first().reset_index()


top_in_each_city[['City', 'Restaurant Name', 'Aggregate rating', 'Votes']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

clean_df = df_merged.dropna(subset=['Aggregate rating', 'Votes'])

plt.figure(figsize=(12,6))
sns.barplot(data=clean_df, x='Aggregate rating', y='Votes', alpha=0.5)
plt.title("Rating vs Votes")
plt.show()

# Correlation
correlation = clean_df['Aggregate rating'].corr(clean_df['Votes'])
print("Correlation between rating and votes:", correlation)
```
